```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv("data/train.csv")
df.head()
```

```python
list(df.columns.values)
```

```python
# Drop columns with null (NaN) values which doesn't add much to analysis
df= df.drop(['Ticket','Cabin'], axis=1) # axis=1 denotes that we refering to columns only and not rows
```

```python
df.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
sns.factorplot(x='Sex', data=df, kind="count")
```

```python
sns.factorplot(x="Sex", data= df, hue="Pclass",kind="count")
```

```python
sns.factorplot(x="Pclass", data=df, kind="count", hue="Sex")
```

```python
## WOW!! So it was interesting thing to find that there are more males in 3rd Pclass. Maybe we need to differentiate between male, female and children

def male_female_child(passenger):
    Age, Sex = passenger
    if Age < 16:
        return 'child'
    else:
        return Sex
    
### Create one more column Person
df['Person'] = df[['Age','Sex']].apply(male_female_child, axis=1)
df[0:10]
```

```python
sns.factorplot(x="Pclass",kind='count',hue='Person',data=df)
```
